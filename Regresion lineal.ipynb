```python
import matplotlib.pyplot as plt
import numpy as np
#from scipy import stats

x = [1.0, 1.6, 3.4, 4.0, 5.2]
y = [1.2, 2.0, 2.4, 3.5, 3.5]

np_x = np.array(x)
np_y = np.array(y)
n = 5

sumaXi = np.sum(np_x)
sumaYi = np.sum(np_y)
sumaXiYi = np.sum(np_x * np_y)
sumaXi2 = np.sum(np_x * np_x)
sX2 = sumaXi * sumaXi

a = ((n * sumaXiYi) - (sumaXi * sumaYi)) / ((n * sumaXi2) - (sumaXi * sumaXi))
b = ((sumaYi * sumaXi2) - (sumaXiYi * sumaXi)) / ((n * sumaXi2) - (sX2))

print('x: ',x)
print('y: ',y)
print('a: ',a)
print('b: ',b)
print('sumaXi: ',sumaXi)
print('sumaYi: ',sumaYi)
print('sumaXiYi: ',sumaXiYi)
print('sumaXi2: ',sumaXi2)

#slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return (b * x) + a

mymodel = list(map(myfunc, x))

plt.grid()
plt.scatter(np_x, np_y)
plt.plot(np_x, mymodel)
plt.show()
```
